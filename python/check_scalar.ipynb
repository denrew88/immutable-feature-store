```python
from fs.config import ScalarShardBuildOptions
from fs.scalar import ScalarDatasetBuilder, write_feature_meta, write_sample_meta
from pathlib import Path
import os
import polars as pl

out_dir = '../data/builder_test'
shard_dir = Path(out_dir) / 'feature_shards'
shard_files = os.listdir(shard_dir)

shard = pl.read_parquet(shard_dir / shard_files[0])


```

```python
from fs.scalar import synthetic

config = synthetic.SyntheticConfig(
    n_samples=5000,
    n_features=16000,
    y_cols=['y1', 'y2', 'y3'],
    seed=100
)


data = synthetic.generate_synthetic(config)
```

```python

out_dir = '../data/builder_test'

sample_rows = []
for i in range(config.n_samples):
    sample_rows.append({
        'sample_key': f'sample_{i:06d}', 'y1': data['Y'][i, 0], 'y2': data['Y'][i, 1], 'y3': data['Y'][i, 2]
    })

sample_meta_path = write_sample_meta(
   sample_rows,
    os.path.join(out_dir, "scalar_sample_meta.parquet"),
)
```

```python
import numpy as np
rng = np.random.default_rng(100)

feature_keys = [f'feature_{i:06d}' for i in range(config.n_features)]
feature_groups = rng.choice(['group_a', 'group_b', 'group_c'], size=config.n_features, replace=True)

feature_rows = [
    {'feature_key': feature_key, 'group': group} 
     for feature_key, group in zip(feature_keys, feature_groups)
]


feature_meta_path = write_feature_meta(
    feature_rows,
    os.path.join(out_dir, "scalar_feature_meta.parquet"),
)

```

```python

builder = ScalarDatasetBuilder(
    out_dir=out_dir,
    sample_meta_path=sample_meta_path,
    feature_meta_path=feature_meta_path,
    build_options=ScalarShardBuildOptions(
        target_shard_mb=4,
        stats_y_cols=config.y_cols,
    ),
)

for i in range(config.n_samples):
    sample_data = {
        feature_key:val
        for feature_key, val in zip(feature_keys, data['X'][i, :])
    }
    builder.write_sample(i, sample_data)



manifest_path = builder.build_shards(keep_sample_major=True)

```

```python
import numpy as np
rng = np.random.default_rng(100)

query_feature_ids = rng.permutation(config.n_features)[:1000]
query_sample_ids = rng.permutation(config.n_samples)[:50]
```

```python
from fs.scalar.reader import open_shard
from tqdm.notebook import tqdm

with open_shard(manifest_path) as ds:
    for fid in tqdm(query_feature_ids):
        batch = ds.get_values(feature_id=fid, sample_ids=query_sample_ids)
        batch_values = np.array([x.value for x in batch.values], dtype=np.float64)
        batch_mask = np.array([x.present for x in batch.values], dtype=bool)

        if not np.array_equal(data['X'][query_sample_ids, fid], batch_values, equal_nan=True):
            print(fid)
```

```python


with open_shard(manifest_path) as ds:
    for batch in tqdm(ds.iter_many(query_feature_ids, query_sample_ids, maintain_order=False), total=query_feature_ids.size):
        fid = batch.feature_id
        batch_values = np.array([x.value for x in batch.values], dtype=np.float64)
        batch_mask = np.array([x.present for x in batch.values], dtype=bool)

        if not np.array_equal(data['X'][query_sample_ids, fid], batch_values, equal_nan=True):
            print(fid)
        
```
